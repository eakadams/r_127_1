Having looked at the VLA data, I did not see anything. I am now convinced that this signal was artifical, based on its narrowness. So I will move forward to calculating upper limits for stellar detection over a reasonable velocity range. First, I need to do the baseline subtraction. Just copy what I did before for easiness.

```python
fig=plt.figure(figsize=(10,6))
f1=fig.add_axes([0.075,0.075,0.85,0.85])
f1.plot(r127['vlsrad'],r127_box31,label='All, 9km/s')
f1.plot(r127_drop['vlsrad'],r127_drop_box31,label='dropped, 9 km/s')
f1.plot(r127['vlsrad'],r127_box83,label='All, 25 km/s')
f1.axis([-1000,4000,0.01,0.04])
f1.legend(loc=2)

f1.set_title('R-127-1')

f1.set_ylabel('Flux [Jy]')
f1.set_xlabel('Vrad,lsr [km/s]')

#find regions to use for baseline subtraction
f1.plot([r127['vlsrad'][61500],r127['vlsrad'][61500]],[-1,10],'k:') #4500 km/s
f1.plot([r127['vlsrad'][60000],r127['vlsrad'][60000]],[-1,10],'k:') #4900 km/s

print(r127['vlsrad'][61500],r127['vlsrad'][60000])

f1.plot([r127['vlsrad'][58500],r127['vlsrad'][58500]],[-1,10],'k:') #5800 km/s
f1.plot([r127['vlsrad'][52000],r127['vlsrad'][52000]],[-1,10],'k:') #5900 km/s

print(r127['vlsrad'][58500], r127['vlsrad'][52000])

f1.plot([r127['vlsrad'][51500],r127['vlsrad'][51500]],[-1,10],'k:') #7200 km/s
f1.plot([r127['vlsrad'][50000],r127['vlsrad'][50000]],[-1,10],'k:') #6500 km/s
#f1.plot([dgsat['vlsrad'][36500],dgsat['vlsrad'][36500]],[-1,10],'k:') #6400 km/s

print(r127['vlsrad'][51500], r127['vlsrad'][50000])
```

```python
#now fit the baselines
#first create arrays just for the fitting:
xfit=np.concatenate((r127['vlsrad'][50000:51500],
                     r127['vlsrad'][52000:58500],
                     r127['vlsrad'][60000:61500]
                     ))

yfit_all=np.concatenate((r127['flux'][50000:51500],
                         r127['flux'][52000:58500],
                         r127['flux'][60000:61500]
                         ))
                        
yfit_drop=np.concatenate((r127_drop['flux'][50000:51500],
                          r127_drop['flux'][52000:58500],
                          r127_drop['flux'][60000:61500]))




#now fit them 
poly_all = np.polyfit(xfit,yfit_all,5)
poly_drop = np.polyfit(xfit,yfit_drop,5)
print(poly_all, poly_drop)

#calculate the fits and plot with the data to check
pol_all_fit = np.polyval(poly_all,r127['vlsrad'])
pol_drop_fit = np.polyval(poly_drop,r127_drop['vlsrad'])


fig=plt.figure(figsize=(10,6))
f1=fig.add_axes([0.075,0.075,0.85,0.85])
f1.plot(r127['vlsrad'],r127_box31,label='All')
f1.plot(r127_drop['vlsrad'],r127_drop_box31,label='dropped')
f1.plot(r127['vlsrad'],pol_all_fit,label='All fit',linewidth=3)
f1.plot(r127_drop['vlsrad'],pol_drop_fit,label='Drop fit',linewidth=3)
f1.plot(r127['vlsrad'],r127_box83,label='All')
f1.plot(r127['vlsrad'],r127_drop_box83,label='drop')
f1.axis([-1000,4000,0.01,0.04])
f1.legend(loc=2)

f1.set_ylabel('Flux [Jy]')
f1.set_xlabel('Vrad,lsr [km/s]')
```

```python
#okay, think that fit is good enough
#at least for now
#so do the baseline subtraction
r127_baselined = r127['flux'] - pol_all_fit
r127_drop_baselined = r127_drop['flux'] - pol_drop_fit

#plot to check how well lined-up the two spectra are as a check
fig=plt.figure(figsize=(10,6))
f1=fig.add_axes([0.075,0.075,0.85,0.85])

f1.plot(r127_drop['vlsrad'],r127_drop_baselined,label='dropped')
f1.plot(r127['vlsrad'],r127_baselined,label='All')
f1.axis([-500,3000,-0.03,0.03])
f1.legend(loc=2)

f1.set_ylabel('Flux [Jy]')
f1.set_xlabel('Vrad,lsr [km/s]')


f1.plot([-1e5,1e5],[0,0],'k--')
```

```python
# boxcar smooth baselined spec

r127_9kms = convolve(r127_baselined, Box1DKernel(30))  #boxcar 30 for ~9 km/s resolution
#r127_drop_box31 = convolve(r127_drop['flux'], Box1DKernel(31))


r127_5kms = convolve(r127_baselined, Box1DKernel(17))
r127_7kms = convolve(r127_baselined, Box1DKernel(24))

r127_25kms = convolve(r127_baselined, Box1DKernel(83))  #boxcar 83 for ~25 km/s resolution
#r127_drop_box83 = convolve(r127_drop['flux'], Box1DKernel(83))
```

```python
fig=plt.figure(figsize=(10,6))
f1=fig.add_axes([0.075,0.075,0.85,0.85])
f1.plot(r127['vlsrad'],r127_9kms,label='9 km/s')
f1.plot(r127['vlsrad'],r127_25kms,label='25 km/s')
#f1.plot(r127['vlsrad'],r127_box83,label='All, 25 km/s',c=cols[2])
f1.axis([-500,1500,-0.005,0.01])
f1.legend(loc=2)

f1.set_title('R-127-1')

f1.set_ylabel('Flux [Jy]')
f1.set_xlabel('Vrad,lsr [km/s]')

#find velocity range over which I will want to calculate things and plot it for reference
ind_150 = np.argmin( np.abs(r127['vlsrad']-150))
ind_1250 = np.argmin( np.abs(r127['vlsrad']-1250))
print(ind_150,ind_1250)

f1.plot([r127['vlsrad'][ind_150],r127['vlsrad'][ind_150]],[-1,10],'k:') #150 km/s
f1.plot([r127['vlsrad'][ind_1250],r127['vlsrad'][ind_1250]],[-1,10],'k:') #1250 km/s
```

```python
#get noise of full res spec over this range
print(np.std(r127_baselined[ind_1250:ind_150]))
print(np.std(r127_drop_baselined[ind_1250:ind_150]))

# all has lower rms - this makes sense because it's more data averaged together, 
# and 1380 shouldn't impact htis portion of spectrum
```

```python
# now do binning of spec - following what I did before
# go for 10, 25, 50 km/s
# will use 50 as a very conservative upper limit - can adapt when know stellar mass and such of gal

dv = abs(r127['vlsrad'][ind_1250]-r127['vlsrad'][ind_1250+1])

width25 = int(25/dv)  #get bin width in channels
width50 = int(50/dv)
width10 = int(10/dv)

print(width25, width50, width10)

#now do the binning, using info found here:
#http://stackoverflow.com/questions/21921178/binning-a-numpy-array

flux=r127_baselined
fluxdrop = r127_drop_baselined

vel = r127['vlsrad']

fluxbin25 = flux[:(flux.size // width25) * width25].reshape(-1, width25).mean(axis=1)
fluxdropbin25 = fluxdrop[:(fluxdrop.size // width25) * width25].reshape(-1, width25).mean(axis=1)

fluxbin50 = flux[:(flux.size // width50) * width50].reshape(-1, width50).mean(axis=1)
fluxdropbin50 = fluxdrop[:(fluxdrop.size // width50) * width50].reshape(-1, width50).mean(axis=1)

fluxbin10 = flux[:(flux.size // width10) * width10].reshape(-1, width10).mean(axis=1)
fluxdropbin10 = fluxdrop[:(fluxdrop.size // width10) * width10].reshape(-1, width10).mean(axis=1)


velbin25 = vel[:(vel.size // width25) * width25].reshape(-1, width25).mean(axis=1)
velbin50 = vel[:(vel.size // width50) * width50].reshape(-1, width50).mean(axis=1)
velbin10 = vel[:(vel.size // width10) * width10].reshape(-1, width10).mean(axis=1)

#also do corresponding boxcar smoothing
flux_box78 = convolve(flux, Box1DKernel(78))
flux_box156 = convolve(flux, Box1DKernel(156))
flux_box31 = convolve(flux, Box1DKernel(31))

flux_drop_box78 = convolve(fluxdrop, Box1DKernel(78))
flux_drop_box156 = convolve(fluxdrop, Box1DKernel(156))
flux_drop_box31 = convolve(fluxdrop, Box1DKernel(31))
```

```python
#things i need for figure

# indices of relevant vel range, for each binned dataset
ind_150_25 = np.argmin( np.abs(velbin25-150))
ind_1250_25 = np.argmin( np.abs(velbin25-1250))

ind_150_50 = np.argmin( np.abs(velbin50-150))
ind_1250_50 = np.argmin( np.abs(velbin50-1250))

ind_150_10 = np.argmin( np.abs(velbin10-150))
ind_1250_10 = np.argmin( np.abs(velbin10-1250))


# rms of flux over that range

rms_flux25 = np.std(fluxbin25[ind_1250_25:ind_150_25])
rms_fluxdrop25 = np.std(fluxdropbin25[ind_1250_25:ind_150_25])

rms_flux50 = np.std(fluxbin50[ind_1250_50:ind_150_50])
rms_fluxdrop50 = np.std(fluxdropbin50[ind_1250_50:ind_150_50])

rms_flux10 = np.std(fluxbin10[ind_1250_10:ind_150_10])
rms_fluxdrop10 = np.std(fluxdropbin10[ind_1250_10:ind_150_10])

# calculate upper limits, assuming W50=50 km/s 
n10 = 5 # nchans for ~50 km/s
n25 = 2
n50 = 1
# need widths of different channel sizes
dv10 = velbin10[ind_1250_10] - velbin10[ind_1250_10+1]
dv25 = velbin25[ind_1250_25] - velbin25[ind_1250_25+1]
dv50 = velbin50[ind_1250_50] - velbin50[ind_1250_50+1]
sint_3sig_10 = 3 * rms_flux10 * n10 * dv10
sint_drop_3sig_10 = 3 * rms_fluxdrop10 * n10 * dv10
sint_3sig_25 = 3 * rms_flux25 * n25 * dv25
sint_drop_3sig_25 = 3 * rms_fluxdrop25 * n25 * dv25
sint_3sig_50 = 3 * rms_flux50 * n50 * dv50
sint_drop_3sig_50 = 3 * rms_fluxdrop50 * n50 * dv50

print(sint_3sig_10, sint_3sig_25, sint_3sig_50)
```

```python
# make a figure showing these different resolutions
fig, ((ax1,ax2) ,(ax3,ax4) ,(ax5,ax6)) = plt.subplots(3, 2,figsize=(14,8),sharex=True,sharey=True)

fig.subplots_adjust(hspace=0,wspace=0)


ax1.axis([-500,1500,-0.0025,0.0025])
#f1.legend(loc=2)

ax1.set_ylabel('Flux [Jy]')
ax3.set_ylabel('Flux [Jy]')
ax5.set_ylabel('Flux [Jy]')
ax3.set_xlabel('Vrad,lsr [km/s]')
ax6.set_xlabel('Vrad,lsr [km/s]')


ax1.plot([velbin10[ind_1250_10],velbin10[ind_1250_10]],[-1,10],'k--')
ax1.plot([velbin10[ind_150_10],velbin10[ind_150_10]],[-1,10],'k--')
ax1.plot([-1e5,1e5],[0,0],'k--')

ax2.plot([velbin10[ind_1250_10],velbin10[ind_1250_10]],[-1,10],'k--')
ax2.plot([velbin10[ind_150_10],velbin10[ind_150_10]],[-1,10],'k--')
ax2.plot([-1e5,1e5],[0,0],'k--')

ax3.plot([velbin25[ind_1250_25],velbin25[ind_1250_25]],[-1,10],'k--')
ax3.plot([velbin25[ind_150_25],velbin25[ind_150_25]],[-1,10],'k--')
ax3.plot([-1e5,1e5],[0,0],'k--')

ax4.plot([velbin25[ind_1250_25],velbin25[ind_1250_25]],[-1,10],'k--')
ax4.plot([velbin25[ind_150_25],velbin25[ind_150_25]],[-1,10],'k--')
ax4.plot([-1e5,1e5],[0,0],'k--')

ax5.plot([velbin50[ind_1250_50],velbin50[ind_1250_50]],[-1,10],'k--')
ax5.plot([velbin50[ind_150_50],velbin50[ind_150_50]],[-1,10],'k--')
ax5.plot([-1e5,1e5],[0,0],'k--')

ax6.plot([velbin50[ind_1250_50],velbin50[ind_1250_50]],[-1,10],'k--')
ax6.plot([velbin50[ind_150_50],velbin50[ind_150_50]],[-1,10],'k--')
ax6.plot([-1e5,1e5],[0,0],'k--')

ax1.bar(velbin10,fluxbin10,label='All, 10 km/s',edgecolor='orange',
       facecolor='none',align='center',width=width10*dv)
ax2.bar(velbin10,fluxdropbin10,label='Drop, 10 km/s',edgecolor='blue',
       facecolor='none',align='center',width=width10*dv)

ax3.bar(velbin25,fluxbin25,label='All, 25 km/s',edgecolor='orange',
       facecolor='none',align='center',width=width25*dv)
ax4.bar(velbin25,fluxdropbin25,label='Drop, 25 km/s',edgecolor='blue',
       facecolor='none',align='center',width=width25*dv)

ax5.bar(velbin50,fluxbin50,label='All, 50 km/s',edgecolor='orange',
       facecolor='none',align='center',width=width50*dv)
ax6.bar(velbin50,fluxdropbin50,label='Drop, 50 km/s',edgecolor='blue',
       facecolor='none',align='center',width=width50*dv)


ax1.set_title('All Subscans')
ax2.set_title('1380 GPS dropped')


# print rms values
ax1.text(200,0.0015,'{1:4.1f} km/s, rms = {0:4.2e} Jy/bm'.format(rms_flux10,dv*width10))
ax2.text(200,0.0015,'{1:4.1f} km/s, rms = {0:4.2e} Jy/bm'.format(rms_fluxdrop10,dv*width10))

ax3.text(200,0.0015,'{1:4.1f} km/s, rms = {0:4.2e} Jy/bm'.format(rms_flux25,dv*width25))
ax4.text(200,0.0015,'{1:4.1f} km/s, rms = {0:4.2e} Jy/bm'.format(rms_fluxdrop25,dv*width25))

ax5.text(200,0.0015,'{1:4.1f} km/s, rms = {0:4.2e} Jy/bm'.format(rms_flux50,dv*width50))
ax6.text(200,0.0015,'{1:4.1f} km/s, rms = {0:4.2e} Jy/bm'.format(rms_fluxdrop50,dv*width50))

ax1.text(200,-0.0015,f'3-$\sigma$ S_int over {n10*dv10:4.1f} km/s is {sint_3sig_10:4.3f} Jy km/s')
ax2.text(200,-0.0015,f'3-$\sigma$ S_int over {n10*dv10:4.1f} km/s is {sint_drop_3sig_10:4.3f} Jy km/s')

ax3.text(200,-0.0015,f'3-$\sigma$ S_int over {n25*dv25:4.1f} km/s is {sint_3sig_25:4.3f} Jy km/s')
ax4.text(200,-0.0015,f'3-$\sigma$ S_int over {n25*dv25:4.1f} km/s is {sint_drop_3sig_25:4.3f} Jy km/s')

ax5.text(200,-0.0015,f'3-$\sigma$ S_int over {n50*dv50:4.1f} km/s is {sint_3sig_50:4.3f} Jy km/s')
ax6.text(200,-0.0015,f'3-$\sigma$ S_int over {n50*dv50:4.1f} km/s is {sint_drop_3sig_50:4.3f} Jy km/s')

ax1.plot([-1e5,1e5],[2*rms_flux10,2*rms_flux10],color='gray')
ax1.plot([-1e5,1e5],[-2*rms_flux10,-2*rms_flux10],color='gray')

ax2.plot([-1e5,1e5],[2*rms_fluxdrop10,2*rms_fluxdrop10],color='gray')
ax2.plot([-1e5,1e5],[-2*rms_fluxdrop10,-2*rms_fluxdrop10],color='gray')

ax3.plot([-1e5,1e5],[2*rms_flux25,2*rms_flux25],color='gray')
ax3.plot([-1e5,1e5],[-2*rms_flux25,-2*rms_flux25],color='gray')

ax4.plot([-1e5,1e5],[2*rms_fluxdrop25,2*rms_fluxdrop25],color='gray')
ax4.plot([-1e5,1e5],[-2*rms_fluxdrop25,-2*rms_fluxdrop25],color='gray')

ax5.plot([-1e5,1e5],[2*rms_flux50,2*rms_flux50],color='gray')
ax5.plot([-1e5,1e5],[-2*rms_flux50,-2*rms_flux50],color='gray')

ax6.plot([-1e5,1e5],[2*rms_fluxdrop50,2*rms_fluxdrop50],color='gray')
ax6.plot([-1e5,1e5],[-2*rms_fluxdrop50,-2*rms_fluxdrop50],color='gray')


plt.savefig('r127_lowz_noise.pdf')
```
